# NETFLIX Data Visualization

```python
# Importing Libraries
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# Reading CSV data
df = pd.read_csv('netflix_titles.csv')
df.head(5)


#Cleaning Data
df = df.dropna(subset=['type' , 'release_year' , 'rating' , 'country' , 'duration'])

#Counting Type Value from Dataset 
type_clount = df['type'].value_counts()
#print(type_clount)



# Diffrenciate Between Movies And Tv shows
plt.Figure(figsize=(6,4))
plt.bar(type_clount.index , type_clount.values , color=["DarkRed" , "crimson"])
plt.title("Numbers of Movies VS TV Shows on Netflix")
plt.xlabel("Types")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig('movies_vs_tv_showes.png')
plt.show()
```

```python
# Counting rating values from dataset
rating_count = df['rating'].value_counts()
#print(rating_count)


# Percentage of rating values from dataset
plt.figure(figsize=(8,6))
plt.pie(rating_count, autopct='%1.1f%%' , labels=rating_count.index ,startangle=50)
plt.title("Percentage of rating")
plt.tight_layout()
plt.savefig('Percentage_of_rating.png')
plt.show()
```

```python
# Extracting movies data from types and copying it 
movies_df =df[df['type'] == 'Movie'].copy()

# Remove Extar spaces and store it into int 
movies_df['duration_int'] = movies_df['duration'].str.replace(' min' , '').astype(int)

# Distribution of mobies
plt.Figure(figsize=(50,12))
plt.hist(movies_df['duration_int'] , bins=30 , color="Maroon" , edgecolor='white')
plt.title("Distribution of movie duration")
plt.xlabel("Duration (minuts)")
plt.ylabel("Numer of movies")
plt.tight_layout()
plt.savefig('Distribution_of_movie_duration.PNG')
plt.show()
```

```python
# Counting and Sorting Releasing year from dataset
releas_count = df['release_year'].value_counts().sort_index()


plt.Figure(figsize=(10,6))
plt.plot(releas_count , releas_count.index , color="darkRed")
plt.title("Release year Vs number of Show")
plt.xlabel("Release Year")
plt.ylabel("Number of Show")
plt.tight_layout()
plt.savefig('Release_year_Vs_number_of_Show.PNG')
plt.show()
```

```python
# Counting Countries Whis hai top numbers of shows
country_show_count = df['country'].value_counts()

# Plotting Bar chart for the Visualizsation
plt.Figure(figsize=(8,6))
plt.bar(country_show_count.index , country_show_count.values, color='teal')
plt.title("Top 10 Country by Number's of show")
plt.xlabel("Number of Shows")
plt.ylabel("Country")
plt.tight_layout()
#plt.savefig('Release_year_Vs_number_of_Show.PNG')
plt.show()
```
